```python
# Convert both to timezone-naive datetime
original['blood_culture_order_datetime'] = pd.to_datetime(original['blood_culture_order_datetime']).dt.tz_localize(None)
test_imp['blood_culture_order_datetime_utc'] = pd.to_datetime(test_imp['blood_culture_order_datetime_utc']).dt.tz_localize(None)

# Now merge
pd.merge(original, test_imp, 
         left_on=["anon_id", "pat_enc_csn_id_coded", "order_proc_id_coded", ], 
         right_on=["anon_id", "pat_enc_csn_id_coded", "order_proc_id_coded",], 
         how="inner")
```

```python
# examine = pd.read_csv("../data/EHR_IncludeInprompt_True/rerun_null_rows/Test_set_df_note_with_predictions_final.csv")
examine = pd.read_csv("../data/EHR_IncludeInprompt_True_new/Test_set_df_note_backup_new.csv")
def parse_classification(json_str):
    try:
        result = json.loads(json_str)
        # return result.get('classification', 'Unknown'), result.get('confidence', 'Unknown'), result.get("verbatim", "Unknown"), result.get("rationale", "Unknown")

        return result.get('Classification', 'Unknown'), result.get('Confidence', 'Unknown'), result.get("Verbatim", "Unknown"), result.get("Rationale", "Unknown")
    except:
        return 'Error', 'Error', 'Error', 'Error'

# Apply parsing to get structured results
classified_data = examine[examine['llm_prediction_result'].notnull()].copy()
classified_data[['classification', 'confidence', 'verbatim', 'rationale']] = classified_data['llm_prediction_result'].apply(
    lambda x: pd.Series(parse_classification(x))
)

# View parsed results
classified_data[['anon_id', 'classification', 'confidence', 'verbatim', 'rationale',"Label"]][["classification", "Label"]].value_counts()
```

```python
# Create prediction column: 1 if classification is INTERMEDIATE or HIGH, 0 otherwise
classified_data['prediction'] = classified_data['classification'].apply(
    # lambda x: 1 if x in ['INTERMEDIATE', 'HIGH'] else 0
    lambda x: 1 if x in ['HIGH'] else 0
)

# Calculate confusion matrix metrics
from sklearn.metrics import confusion_matrix

# Filter to rows where both Label and prediction are available
valid_data = classified_data[classified_data['Label'].notnull() & classified_data['prediction'].notnull()].copy()

if len(valid_data) > 0:
    # Get true labels and predictions
    y_true = valid_data['Label'].astype(int)
    y_pred = valid_data['prediction'].astype(int)
    
    # Calculate confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    
    # Extract values from confusion matrix
    tn, fp, fn, tp = cm.ravel()
    
    # Calculate metrics
    sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
    ppv = tp / (tp + fp) if (tp + fp) > 0 else 0
    npv = tn / (tn + fn) if (tn + fn) > 0 else 0
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    
    print(f"Sensitivity: {sensitivity:.4f}")
    print(f"Specificity: {specificity:.4f}")
    print(f"PPV:         {ppv:.4f}")
    print(f"NPV:         {npv:.4f}")
    print(f"Accuracy:    {accuracy:.4f}")
    
else:
    print("No valid data found with both Label and prediction columns!")
```

```python
# Balanced 50-50 sampling for confusion matrix metrics
from sklearn.metrics import confusion_matrix
import numpy as np

# Filter to rows where both Label and prediction are available
valid_data = classified_data[classified_data['Label'].notnull() & classified_data['prediction'].notnull()].copy()

if len(valid_data) > 0:
    # Separate positive and negative cases
    positive_cases = valid_data[valid_data['Label'] == 1]
    negative_cases = valid_data[valid_data['Label'] == 0]
    
    print(f"Original data: {len(positive_cases)} positive, {len(negative_cases)} negative")
    
    # Calculate sample sizes for 50-50 split
    min_class_size = min(len(positive_cases), len(negative_cases))
    sample_size_per_class = min_class_size  # Use all of the smaller class
    
    # Randomly sample equal numbers from each class
    # np.random.seed(42)  # For reproducibility
    sampled_positive = positive_cases.sample(n=sample_size_per_class, random_state=42)
    sampled_negative = negative_cases.sample(n=sample_size_per_class)
    
    # Combine the balanced sample
    balanced_data = pd.concat([sampled_positive, sampled_negative])
    
    print(f"Balanced sample: {len(sampled_positive)} positive, {len(sampled_negative)} negative")
    
    # Get true labels and predictions from balanced sample
    y_true = balanced_data['Label'].astype(int)
    y_pred = balanced_data['prediction'].astype(int)
    
    # Calculate confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    
    # Extract values from confusion matrix
    tn, fp, fn, tp = cm.ravel()
    
    # Calculate metrics
    sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
    ppv = tp / (tp + fp) if (tp + fp) > 0 else 0
    npv = tn / (tn + fn) if (tn + fn) > 0 else 0
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    
    print(f"\nBalanced Sample Metrics:")
    print(f"Sensitivity: {sensitivity:.4f}")
    print(f"Specificity: {specificity:.4f}")
    print(f"PPV:         {ppv:.4f}")
    print(f"NPV:         {npv:.4f}")
    print(f"Accuracy:    {accuracy:.4f}")
    
else:
    print("No valid data found with both Label and prediction columns!")
```
